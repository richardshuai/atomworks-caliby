# Setup & Imports

```python
from cifutils.utils.io_utils import load_any, to_cif_file
from cifutils.utils.visualize import view
from biotite.database.rcsb import fetch
```

While `cifutils` is called `cifutils`, it can actually load many more file formats than just cif files (among them: `.pdb`, `.cif`, `.cif.gz`, `.bcif`, `.bcif.gz`, and buffers thereof).

(But YES! You should use `.cif` files instead of `.pdb` files whenever you can! They have many advantages -- and the support for them in `cifutils` is much better!)

```python
# Load from a pdb
load_any(fetch("6lyz", "pdb"), model=1)
```

```python
load_any(
    "/databases/rcsb/cif/oc/5ocm.cif.gz",
    model=1,
    extra_fields=["occupancy", "atom_id", "b_factor", "charge"],
    include_bonds=True,
)
```

```python
# We can also parse files
from cifutils import CIFParser

data = CIFParser().parse(
    "/databases/rcsb/cif/oc/5ocm.cif.gz",
    extra_fields=["occupancy", "atom_id", "b_factor", "charge"],
    include_bonds=True,
)
```

```python
from cifutils.utils.visualize import view

view(data["assemblies"]["1"][0])
```

```python
# You can then save it out again
to_cif_file(data["assemblies"]["1"][0], "./5ocm_mod.cif", id="5ocm_modified")
```
